```python
with open('store_names.txt', 'r') as file:
    stores_names = file.read().splitlines()

#creating the key names for all the stores and putting them in categories
key_names = { "grocery": ['walmart', 'target', 'costco', 'sams club',"sam's club", 'kroger', 'walgreens', 'cvs', 'meijer', 
                           'aldi', 'whole foods', 'trader joes', "trader joe's","publix", "safeway", "albertsons", "schnucks",
                           "jewel-osco", "jewel osco", "hy-vee", "hyvee","county market", "fresh thyme", "green top grocery", "save a lot"],
               
               "home": ['home depot', 'lowes', "lowe's", "menards"],
               
               "food": ["mcdonalds", "mcdonald's", "chick-fil-a", "taco bell", "wendys", "wendy's", "burger king",
                        "arby's", "arbys", "kfc", "popeyes", "raising canes", "raising cane's", "culvers", "culver's", "portillos", "portillo's"
                        "starbucks", "dunkin", "dunkin donuts","dairy queen", "sonic", "five guys","shake shack", "wingstop"
                        "subway", "jimmy johns", "jimmy john's","jersey mikes", "jersey mike's","firehouse subs", "potbelly", "panera",
                        "chipotle", "qdoba", "panda express","olive garden", "applebees", "applebee's","chilis", "chili's", "buffalo wild wings",
                        "bdubs","red robin", "outback", "texas roadhouse", "cracker barrel", "ihop", "dennys", "denny's","bob evans", "waffle house"],
               
               "retail": ["best buy", "hobby lobby", "von maur", "tj maxx", "tjmaxx", "marshalls", "ross", "kohls", "kohl's", "jcpenney",
                         "jc penney", "bed bath beyond", "ulta","bath body works", "old navy", "gap","american eagle", "rue 21", "rue21",
                        "gamestop", "five below", "big lots","staples", "office depot", "sephora"],
                
                "gas": ["autozone", "o'reilly", "oreilly", "advance auto", "napa", "shell", "bp",
                          "marathon", "speedway", "casey's", "caseys", "circle k", "mobil", "exxon", "phillips 66"]}

receipt_dict = {}
#creating category dictionaries that will put the receipt_data into these dictonaries based on key names
grocery_dict = {"grocery":[]}
home_dict = {"home":[]}
food_dict = {"food":[]}
retail_dict = {"retail":[]}
gas_dict = {"gas":[]}
#initializing the totals for each category
grocery_total = 0
home_total = 0
food_total = 0
retail_total = 0
gas_total = 0
for i,line in enumerate(open("receipt_data.txt","r")):
    line = line.strip()
    parts = line.split(", ")
    receipt_store = parts[0].lower()
    date = parts[1]
    price = parts[2]
    receipt_dict = {receipt_store: [date, price]}
    #is searching for which category the receipt_dict matches with and then appends the category
    for category, stores in key_names.items():
        if receipt_store in stores:
            receipt_dict[receipt_store].append(category)
            #is appending recieipt dictionaries into each category dictionary
            if category == "grocery":
                grocery_dict["grocery"].append(receipt_dict)
            elif category == "home":
                home_dict["home"].append(receipt_dict)
            elif category == "food":
                food_dict["food"].append(receipt_dict)
            elif category == "retail":
                retail_dict["retail"].append(receipt_dict)
            elif category == "gas":
                gas_dict["gas"].append(receipt_dict)                
#function used to get the totals per category     
def get_totals (dictionary,key,total):
    for receipt_dict in dictionary[key]:
        for receipt_store in receipt_dict:
            data = receipt_dict[receipt_store]
            price = float(data[1])
            total += price
    return total

print(get_totals(grocery_dict,'grocery', grocery_total))
print(get_totals(home_dict, 'home', home_total))
print(get_totals(food_dict, 'food', food_total))
print(get_totals(retail_dict, 'retail', retail_total))
print(get_totals(gas_dict, 'gas', gas_total))
```
